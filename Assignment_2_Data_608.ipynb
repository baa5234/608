```python

import pandas as pd
import matplotlib.pyplot as plt


cpi_df = 'https://raw.githubusercontent.com/baa5234/608/refs/heads/main/assignment%202/CPI.csv'
fedfunds_df = 'https://raw.githubusercontent.com/baa5234/608/refs/heads/main/assignment%202/fedfunds.csv'
unemployment_df = 'https://raw.githubusercontent.com/baa5234/608/refs/heads/main/assignment%202/unemployment%20rate%20.csv'

cpi_df = pd.read_csv(cpi_df)
fedfunds_df = pd.read_csv(fedfunds_df)
unemployment_df = pd.read_csv(unemployment_df)

print(cpi_df.head())
print(fedfunds_df.head())
print(unemployment_df.head())

```

```python
# Convert observation_date to datetime
fedfunds_df['observation_date'] = pd.to_datetime(fedfunds_df['observation_date'])

# Extract year and month
fedfunds_df['Year'] = fedfunds_df['observation_date'].dt.year
fedfunds_df['Month'] = fedfunds_df['observation_date'].dt.strftime('%b')  # Abbreviated month names

# Pivot the DataFrame to get the required format
pivot_fedfunds = fedfunds_df.pivot(index='Year', columns='Month', values='FEDFUNDS')

# Ensure correct column order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pivot_fedfunds = pivot_fedfunds[month_order]

# Reset index to move 'Year' back to a column
pivot_fedfunds.reset_index(inplace=True)

# Print or save the formatted DataFrame
print(pivot_fedfunds.head())
```

```python


# Removing 'HALF1' and 'HALF2' are columns in cpi_df
if 'HALF1' in cpi_df.columns and 'HALF2' in cpi_df.columns:
    cpi_df = cpi_df.drop(['HALF1', 'HALF2'], axis=1)
else:
    print("Columns 'HALF1' and/or 'HALF2' not found in cpi_df.")

print(cpi_df.head())
print(fedfunds_df.head())
print(unemployment_df.head())


# Convert observation_date to datetime
fedfunds_df['observation_date'] = pd.to_datetime(fedfunds_df['observation_date'])

# Extract year and month
fedfunds_df['Year'] = fedfunds_df['observation_date'].dt.year
fedfunds_df['Month'] = fedfunds_df['observation_date'].dt.strftime('%b')  # Abbreviated month names

# Pivot the DataFrame to get the required format
pivot_fedfunds = fedfunds_df.pivot(index='Year', columns='Month', values='FEDFUNDS')

# Ensure correct column order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pivot_fedfunds = pivot_fedfunds[month_order]

# Reset index to move 'Year' back to a column
pivot_fedfunds.reset_index(inplace=True)


```

```python


# Remove rows where 'Year' is 2025 (if the year column exists in the dataframe)
for df_name, df in zip(["pivot_fedfunds", "cpi_df", "unemployment_df"], [pivot_fedfunds, cpi_df, unemployment_df]):
    if "Year" in df.columns:
        df.drop(df[df["Year"] == 2025].index, inplace=True)
    else:
        print(f"Column 'Year' not found in {df_name} DataFrame.")

print(pivot_fedfunds.tail())
print(cpi_df.tail())
print(unemployment_df.tail())
```

```python
unemployment_melted = unemployment_df.melt(id_vars="Year", var_name="Month", value_name="Unemployment Rate")
cpi_melted = cpi_df.melt(id_vars="Year", var_name="Month", value_name="CPI")
fedfunds_melted = pivot_fedfunds.melt(id_vars="Year", var_name="Month", value_name="Fed Funds Rate")

# Merge all dataframes on 'Year' and 'Month'
merged_df = pd.merge(unemployment_melted, cpi_melted, on=["Year", "Month"])
merged_df = pd.merge(merged_df, fedfunds_melted, on=["Year", "Month"])

# Group by Year and calculate the mean for each year
yearly_avg = merged_df.groupby("Year").mean(numeric_only=True)

# Plot the data
plt.figure(figsize=(10, 4))
plt.plot(yearly_avg.index, yearly_avg["Unemployment Rate"], label="Unemployment Rate", marker="o", linestyle="-")
plt.plot(yearly_avg.index, yearly_avg["Fed Funds Rate"], label="Fed Funds Rate", marker="^", linestyle="-")


# Formatting the plot
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.title("Yearly Averages of Unemployment Rate, and Fed Funds Rate")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

This is the Fed Funds Rate (Interest Rate) vs. Unemployment Rate over time.

This graph shows:

* If the Fed Fund raises rates, we expect higher unemployment due to economic slowdowns.
* If the Fed Fund lowers rates, we expect lower unemployment as borrowing and investment increase.

The line graph offers insight into how the Federal Reserve's interest rate policies align with unemployment trends, **but it alone does not fully determine whether the Fed has met its congressional mandate.** While stable unemployment amid fluctuating rates suggests economic balance, external factors like recessions and global events also play a role.

```python
# Converting dataframes to long format and compute yearly averages
cpi_melted = cpi_df.melt(id_vars="Year", var_name="Month", value_name="CPI").groupby("Year").mean(numeric_only=True)
fedfunds_melted = pivot_fedfunds.melt(id_vars="Year", var_name="Month", value_name="Fed Funds Rate").groupby("Year").mean(numeric_only=True)
unemployment_melted = unemployment_df.melt(id_vars="Year", var_name="Month", value_name="Unemployment Rate").groupby("Year").mean(numeric_only=True)

# Merging all data into a single DataFrame
merged_df = cpi_melted.merge(fedfunds_melted, on="Year").merge(unemployment_melted, on="Year")

# Plot Inflation vs. Interest Rates
plt.figure(figsize=(10, 4))
plt.plot(merged_df.index, merged_df["CPI"], label="CPI (Inflation)", color="red", marker="s", linestyle="-")
plt.plot(merged_df.index, merged_df["Fed Funds Rate"], label="Fed Funds Rate (Interest Rate)", color="blue", marker="o", linestyle="-")

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Inflation vs. Interest Rates")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

```

This graph shows:


*   Lower inflation with higher
interest rates suggests effective Fed control
*   Persistent inflation despite rate hikes indicates external factors at play.
* Falling rates with rising inflation suggest expansionary policy effects.



The Inflation vs. Interest Rates graph partially answers the Fed’s mandate by showing how well it manages price stability through interest rate adjustments. However, it does not address the employment mandate, which is equally important for evaluating the Fed’s overall effectiveness.

```python
# Scatter Plot: Interest Rate vs. Unemployment Rate
plt.figure(figsize=(6, 4))
plt.scatter(merged_df["Fed Funds Rate"], merged_df["Unemployment Rate"], color="green", alpha=0.7)
plt.xlabel("Fed Funds Rate (Interest Rate)")
plt.ylabel("Unemployment Rate")
plt.title("Interest Rate vs. Unemployment Rate")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Scatter Plot: CPI vs. Fed Funds Rate
plt.figure(figsize=(6, 4))
plt.scatter(merged_df["Fed Funds Rate"], merged_df["CPI"], color="purple", alpha=0.7)
plt.xlabel("Fed Funds Rate (Interest Rate)")
plt.ylabel("CPI (Inflation)")
plt.title("CPI vs. Fed Funds Rate")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### It seems as though most of these graph did not answer the question therefore I will do a graph with a benchmark of 2%.

```python
economic_turning_points = {
    2001: "Tech Market Crash",
    2008: "Great Reccession",
    2020: "COVID-19 Pandemic"
}

plt.figure(figsize=(10, 4))

# Plot CPI (Inflation)
plt.plot(merged_df.index, merged_df["CPI"], label="CPI (Inflation)", color="red", marker="s", linestyle="-")

# Plot Fed Funds Rate
plt.plot(merged_df.index, merged_df["Fed Funds Rate"], label="Fed Funds Rate (Interest Rate)", color="blue", marker="o", linestyle="-")

# Plot Unemployment Rate
plt.plot(merged_df.index, merged_df["Unemployment Rate"], label="Unemployment Rate", color="green", marker="^", linestyle="-")

# Add Benchmark Lines
plt.axhline(y=2, color="red", linestyle="--", label="Fed Inflation Target (2%)")
plt.axhline(y=4.5, color="green", linestyle="--", label="Fed Unemployment Target (~4.5%)")

# Highlight key economic turing points on the graph
for year, event in economic_turning_points.items():
    plt.axvline(x=year, color="gray", linestyle="--", alpha=0.6)
    plt.text(year, plt.ylim()[1] * 0.9, event, rotation=90, verticalalignment="top", fontsize=10, color="black")

# Formatting the plot
plt.xlabel("Year")
plt.ylabel("Rate/Index")
plt.title("Assessing the Fed's Mandate: Inflation, Interest Rates, and Unemployment Over Time with Economic Events")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

This detailed visualization shows how well the Federal Reserve successfully balances inflation and employment over time.

This graph shows:

* If inflation stays near 2% and unemployment around 4.5%, the Fed is meeting its goal.
* A spike in unemployment after rate hikes suggests that higher interest rates are slowing job growth.
* If inflation remains high despite rate increases, other factor may be driving it.
